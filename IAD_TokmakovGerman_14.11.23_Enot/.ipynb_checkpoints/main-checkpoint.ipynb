```python
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np

# first image
plt.subplot(2, 2, 1)
face = misc.face(gray=True)
plt.imshow(face)

# second image
plt.subplot(2, 2, 2)
face = misc.face(gray=True)
plt.imshow(face, cmap='gray')

# third image
plt.subplot(2, 2, 3)
crop_face = face[100:-100, 100:-100]
plt.imshow(crop_face, cmap='gray')

# fourth image
plt.subplot(2, 2, 4)
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]

y.shape, x.shape

centerx, centery = (650, 250)

mask = ((x - centerx) ** 2 + (y - centery) ** 2) > 250 ** 2
face[mask] = 0
plt.imshow(face, cmap='gray')

# show
plt.show()



```

```python
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np

# first image
plt.subplot(2, 2, 1)
face = misc.face(gray=True)
plt.imshow(face)

# second image
plt.subplot(2, 2, 2)
face = misc.face(gray=True)
plt.imshow(face, cmap='gray')

# third image
plt.subplot(2, 2, 3)
crop_face = face[100:-100, 100:-100]
plt.imshow(crop_face, cmap='gray')

# fourth image
plt.subplot(2, 2, 4)
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]

y.shape, x.shape

centerx, centery = (650, 270)

a = 300
b = 200

mask = ((x - centerx) ** 2 / (a ** 2) + (y - centery) ** 2 / (b ** 2)) > 1
face[mask] = 0
plt.imshow(face, cmap='gray')

# show
plt.show()
```

```python
import numpy as np
data = np.loadtxt('data/populations.txt')
year, hares, lynxes, carrots = data.T  #   

import matplotlib.pyplot as plt
plt.axes([0.2, 0.1, 0.5, 0.8]) 

plt.plot(year, hares, year, lynxes, year, carrots) 

plt.legend(('Hare', 'Lynx', 'Carrot'), loc=(1.05, 0.5)) 

plt.show()

print("1.Average values: ")
print("Average for hares: ", hares.mean())
print("Average for lynxes: ", lynxes.mean())
print("Average for carrots: ", carrots.mean(), "\n")


print("2.Standard deviation: ")
print("Standard deviation for hares: ", hares.std())
print("Standard deviation for lynxes: ", lynxes.std())
print("Standard deviation for carrots: ", carrots.std(), "\n")


print("3.Max population value: ")
print("Max population value for hares: ", hares.max(), end=" ")
print("And it's year: ", year[list(hares).index(hares.max())])

print("Max population value for lynxes: ", lynxes.max(), end=" ")
print("And it's year: ", year[list(lynxes).index(lynxes.max())])

print("Max population value for carrots: ", carrots.max(), end=" ")
print("And it's year: ", year[list(carrots).index(carrots.max())], "\n")

print("4.Kind of data with max population for all years: ", max(hares.sum(), lynxes.sum(), carrots.sum()), end=" ")
args = np.argsort([hares.sum(), lynxes.sum(), carrots.sum()])
print(["Hares", "Lynxes", "Carrots"][args[-1]])

print("5.Top two years for most lower population level for Hare:")
args = np.argsort(hares)
f_val = hares[args][0]
s_val = hares[args][1]

print(year[list(hares).index(f_val)])
print(year[list(hares).index(s_val)])

print("Top two years for most lower population level for Lynx:")
args = np.argsort(lynxes)
f_val = lynxes[args][0]
s_val = lynxes[args][1]

print(year[list(lynxes).index(f_val)])
print(year[list(lynxes).index(s_val)])

print("Top two years for most lower population level for Carrot:")
args = np.argsort(carrots)
f_val = carrots[args][0]
s_val = carrots[args][1]

print(year[list(carrots).index(f_val)])
print(year[list(carrots).index(s_val)])
print()

print("6.")


plt.subplot(1, 3, 1)
plt.title("Hares gradient")
plt.plot(year, np.gradient(hares))


plt.subplot(1, 3, 2)
plt.title("Lynxes gradient")
plt.plot(year, np.gradient(lynxes))

plt.subplot(1, 3, 3)
plt.title("Lynxes and hares corecoeff")
plt.plot(np.corrcoef(hares, lynxes))

plt.figure(200)
plt.show()



```
