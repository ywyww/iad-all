## 1. Провести кластеризацию набора алгоритмом k-ближайших соседей. Визуализировать результат.

```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap
import numpy as np

iris = datasets.load_iris()

import matplotlib.pyplot as plt

X, y = iris.data, iris.target

X = X[:, :2]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.7, random_state=42
)

classifier = KNeighborsClassifier(3)
cmap_light = ListedColormap(["#FFAAAA", "#AAFFAA", "#AAAAFF"])
cmap_bold = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])

fig, ax = plt.subplots()
classifier = classifier.fit(X_train, y_train)
score = classifier.score(X_test, y_test)

DecisionBoundaryDisplay.from_estimator(
        classifier,
        X,
        cmap=cmap_light,
        alpha=0.8,
        ax=ax,
        response_method="predict",
        plot_method="pcolormesh",
        shading="auto",
    )

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor="k", s=20)


ax.set_xlabel("sepal length")
ax.set_ylabel("sepal width")

plt.show()
```

## 4. Провести кластеризацию набора алгоритмом k-средних. Визуализировать результат.

```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

iris = datasets.load_iris()


X, y = iris.data, iris.target

X = X[:, :2]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.7, random_state=42
)

classifier = KMeans(3)
cmap_light = ListedColormap(["#FFAAAA", "#AAFFAA", "#AAAAFF"])
cmap_bold = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])

fig, ax = plt.subplots()
classifier = classifier.fit(X_train, y_train)
score = classifier.score(X_test, y_test)

DecisionBoundaryDisplay.from_estimator(
        classifier,
        X,
        cmap=cmap_light,
        alpha=0.8,
        ax=ax,
        response_method="predict",
        plot_method="pcolormesh",
        shading="auto",
    )

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor="k", s=20)


ax.set_xlabel("sepal length")
ax.set_ylabel("sepal width")

plt.show()
```

## 5. Провести иерархическую кластеризацию методом Уорда. Визуализировать дендрограмму.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt


iris = datasets.load_iris()


X, y = iris.data, iris.target

X = X[:, :2]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.7, random_state=42
)

fig, ax = plt.subplots()

w = linkage(X_train, "ward")
dendrogram(w)

plt.show()
```

## 6. Провести спектральную кластеризацию. Визуализировать результат.

```python
from sklearn import datasets
from sklearn.cluster import SpectralClustering
from sklearn.model_selection import train_test_split
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap
from matplotlib.pyplot import subplots_adjust
import matplotlib.pyplot as plt

iris = datasets.load_iris()


X, y = iris.data, iris.target

X = X[:, :2]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.7, random_state=42
)

classifier = SpectralClustering(3, affinity="nearest_neighbors", assign_labels="discretize")
cmap_light = ListedColormap(["#FFAAAA", "#AAFFAA", "#AAAAFF"])
cmap_bold = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])

fig, (ax1, ax2,) = plt.subplots(1, 2)
classifier.fit(X_train)

ax1.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor="k", s=20)
ax1.set_xlabel("sepal length")
ax1.set_ylabel("sepal width")
ax1.set_title("Iris set")

ax2.scatter(X[:, 0], X[:, 1], c=classifier.fit_predict(X), edgecolor='k', s=20)
ax2.set_xlabel("sepal length")
ax2.set_ylabel("sepal width")
ax2.set_title("Spectral clustering predict")

plt.subplots_adjust(wspace=0.4)
plt.show()
```
